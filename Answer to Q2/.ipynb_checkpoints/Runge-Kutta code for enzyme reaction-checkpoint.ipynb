```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

### 8.2

```python
'''
C[0],C[1],C[2],C[3] is the concentration of E,S,ES,P,respectively
'''

# define the expression of changing rate of different species
def all_function(C,m):
    if m==0:
        return 750*(C[2])-100*(C[0])*(C[1])
    if m==1:
        return 600*(C[2])-100*(C[0])*(C[1])
    if m==2:
        return 100*(C[0])*(C[1])-750*(C[2])
    if m==3:
        return 150*(C[2])
    
    
```

```python
# define the Runge-Kutta function for the enzyme reaction
def compute_one_step_rk4(C,interval,funct):
    next_C=[]
    for i in range(4):
        var_C=C[:]                     # reset the previous C value
        k1=funct(var_C,i)
        var_C[i]=C[i]+0.5*interval*k1  # update parameter to Yn+(h*k1)/2
        k2=funct(var_C,i)
        var_C[i]=C[i]+0.5*interval*k2  # update parameter to Yn+(h*k2)/2
        k3=funct(var_C,i)
        var_C[i]=C[i]+interval*k3      # update parameter to Yn+h*k3
        k4=funct(var_C,i)
        next_C.append(C[i]+interval*(k1+2*k2+2*k3+k4)/6) # store the next C value
    return next_C
    
```

```python
# set the initial parameters, iteration number and step size 
concentration=[1,10,0,0]
iter_num=5000000
h=0.0000001
time=0                            # start time is 0
result=[(time,concentration)]     # record first parameters


for j in range(iter_num):         # start the iteration
    concentration=compute_one_step_rk4(concentration,h,all_function)
    time=time+h                   
    result.append((time,concentration))
```

```python
t,E,S,ES,P=[],[],[],[],[]

# Save the results to their respective lists
for i in range(len(result)):
    tup=result[i]
    t.append(tup[0])       # tup[0] is time
    E.append(tup[1][0])    # tup[1][0] : C[E]
    S.append(tup[1][1])    # tup[1][1] : C[S]
    ES.append(tup[1][2])    # tup[1][2] : C[ES]
    P.append(tup[1][3])    # tup[1][3] : C[P]
    
    

```

```python
 # Plot the result
plt.figure(figsize=(10,8),dpi=300)
plt.plot(t, E, label='E')
plt.plot(t, S, label='S')
plt.plot(t, ES, label='ES')
plt.plot(t, P, label='P')
plt.xlabel('time')
plt.ylabel('Concentration')

plt.legend()
plt.show()
```

### 8.3

```python
V=[i*150 for i in ES]      # V = Change rate of P = k3 * C[ES] = 150 * C[ES]
```

```python
# Plot the curve of V-C[S]

max_indx=np.argmax(V)       # get max value index
plt.figure(figsize=(10,8),dpi=300)
plt.plot(S, V, label='V=150*C(ES)')
plt.plot(S[max_indx],V[max_indx],'ko')
plt.annotate(text='Max(V)={:.2f} When C[S]={:.2f}'.format(V[max_indx],S[max_indx]),xy=(S[max_indx],V[max_indx]))
plt.xlabel('Concentration of S')
plt.ylabel('Changing rate of P')

plt.legend()
plt.show()
```
